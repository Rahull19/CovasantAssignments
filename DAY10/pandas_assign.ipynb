```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bikes.csv')
df.head()
```

```python
df['DateTime'] = pd.to_datetime(df['Date'], dayfirst=True)
df.set_index('DateTime', inplace=True)
df[['Berri1', 'PierDup']].plot(figsize=(12, 5), title='Daily Attendance - Berri1 & PierDup')
plt.ylabel("Bike Count")
plt.show()
```

```python
df['Weekday'] = df.index.day_name()
print(df['Weekday'].value_counts())
```

```python
weekday_sum = df.groupby('Weekday')[['Berri1', 'PierDup']].sum()
print(weekday_sum)
```

```python
import numpy as np
```

```python
fs = 500
t = np.arange(0, 1, 1/fs)
signal = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t)
```

```python
fft_vals = np.fft.fft(signal)
fft_freqs = np.fft.fftfreq(len(fft_vals), 1/fs)
```

```python
plt.plot(fft_freqs[:len(fft_vals)//2], np.abs(fft_vals[:len(fft_vals)//2]))
plt.title("FFT of 5Hz + 50Hz Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
data1 = np.random.normal(0, 1, 10000)
data2 = np.random.normal(3, 1.5, 10000)

plt.hist(data1, bins=50, alpha=0.5, label='N(0,1)')
plt.hist(data2, bins=50, alpha=0.5, label='N(3,1.5)')
plt.hist(data1 + data2, bins=50, alpha=0.7, label='Sum')
plt.legend()
plt.title("Addition of Two Gaussian Distributions")
plt.show()
```
